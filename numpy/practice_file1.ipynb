```python
import numpy as np
zeroes=np.zeros(10)
print(zeroes)
vowels=np.array(['a','e','i','o','u'])
print(vowels)
ones_2d=np.ones((2,5),dtype=int)
print(ones_2d)
myarray1=np.array([[2.7,-2,-19],[0,3.4,99.9],[10.6,0,13]])
print(myarray1)
myarray2=np.arange(4,61,4,dtype=float).reshape(3,5)
print(myarray2)
```

```python
print("dimension:",zeroes.ndim,"shape:",zeroes.shape,"size:",zeroes.size,"data type:",zeroes.dtype,"itemsize:",zeroes.itemsize)
print("dimension:",vowels.ndim,"shape:",vowels.shape,"size:",vowels.size,"data type:",vowels.dtype,"itemsize:",vowels.itemsize)
print("dimension:",ones_2d.ndim,"shape:",ones_2d.shape,"size:",ones_2d.size,"data type:",ones_2d.dtype,"itemsize:",ones_2d.itemsize)
print("dimension:",myarray1.ndim,"shape:",myarray1.shape,"size:",myarray1.size,"data type:",myarray1.dtype,"itemsize:",myarray1.itemsize)
print("dimension:",myarray2.ndim,"shape:",myarray2.shape,"size:",myarray2.size,"data type:",myarray2.dtype,"itemsize:",myarray2.itemsize)

ones_new=ones_2d.reshape(1,10)
print(ones_new)
print(vowels[[1,2]])
print(myarray1[1:3,])
print(myarray1[:,0:2])
print(myarray1[1:3, 0])
print(vowels[::-1])
```

```python
print(ones_2d/3)
print((myarray1**3)/2)
result=np.sqrt(myarray2)/2
print(np.round(result,2))#result 2 place decimal
myarray3=myarray1@myarray2#matrix multiplication
print(myarray3)
#myarray1(3,3) and myarray2(3,5)
#Due to incompatible shapes element-wise operation can't perform 
#print(myarray1/myarray2)
#print(myarray1*myarray2)
#print(myarray1+myarray2)
#print(myarray2-myarray1)
```

```python
print(ones_2d.transpose())#change rows into columns
print(myarray2.transpose())
sorted=np.sort(vowels)
print(sorted[::-1])#reverse order
sorted2=np.sort(myarray1,axis=0)#column-wise sorted
print(sorted2)
```

```python
myarray2A,myarray2B,myarray2C,myarray2D,myarray2E=np.hsplit(myarray2,5)#horizontal split
myarray2A,myarray2B,myarray2C,myarray2D,myarray2E=np.split(myarray2,5,axis=1)#axis=1 ->column-wise
print(myarray2A,myarray2B,myarray2C,myarray2D,myarray2E)

zeroesA,zeroesB,zeroesC,zeroesD=np.split(zeroes,[2,5,7])
print(zeroesA,zeroesB,zeroesC,zeroesD)

arr=np.concatenate((myarray2A,myarray2B,myarray2C),axis=1)
print(arr)
```

```python
myarray4=np.arange(-1,9.5,0.25).reshape(14,3)
print(myarray4)
a1,a2,a3=np.array_split(myarray4,3,axis=0)#array_split(), works even when division is not exact
print(a1,a2,a3)
```
