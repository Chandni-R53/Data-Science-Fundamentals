- describe(include="all"), includes all columns
- describe(include="object"), includes only categorical columns

```python
data2={
    "Name":["Aliza","Sofiya","Joe","Joe","John"],
    "Age":[23,34,45,22,26],
    "Salary":[38000,45000,25000,67000,54000],
    "Performance":[57,60,80,46,92]
}
dF=pd.DataFrame(data2)
print(dF)
print(dF.describe())
print(dF.describe(include="object"))
print(dF.describe(include="all"))
```

Sorting and Aggregation
- sort_values(), to sort the values in ascending/descending
- aggregation methods like sum(),mean(),std(),min(),max(),count()
- groupby(), to group the columns 

```python
data4={
    "Name":["Fredrik","Tiya","Alex","Olivia","Johnsy"],
    "Age":[23,34,34,23,26],
    "Salary":[58000,67000,25000,67000,54000],
    
}
dF4=pd.DataFrame(data4)
print(dF4)
dF4.sort_values(by='Name',ascending=True,inplace=True)
dF4.sort_values(by=['Age','Salary'],ascending=False,inplace=True)
print(dF4)
#Aggregation
print("Aggregation of data")
print(dF4['Salary'].mean())
print(dF4['Age'].max())
print(dF4['Salary'].min())
#data grouping
group=dF4.groupby('Age')['Salary'].sum()
print(group)
group=dF4.groupby(['Age','Name'])['Salary'].sum()#grouping with multiple columns
print(group)
```

Merging and joining
- pd.merge(), join data by key
- pd.concat(), vertical/horizontal stacking

```python
customers=pd.DataFrame({
    "CustomerID":[1,2,3,4],
    "CustomerName":["Amit","Sumit","Kiran","Pushpa"]
})
orders=pd.DataFrame({
    "CustomerID":[1,2,3,5,6],
    "OrderAmount":[999,580,260,1500,1240]
})
merged=pd.merge(customers,orders,on="CustomerID",how="inner")
print("Inner Join")
print(merged)
merged2=pd.merge(customers,orders,on="CustomerID",how="outer")
print("Outer Join")
print(merged2)
merged3=pd.merge(customers,orders,on="CustomerID",how="left")
print("Left Join")
print(merged3)
merged4=pd.merge(customers,orders,on="CustomerID",how="right")
print("Right Join")
print(merged4)
#Concatenation 
print("Concatenation of dataframes:")
concated=pd.concat([customers,orders],axis=1,ignore_index=True)
print(concated)
```
