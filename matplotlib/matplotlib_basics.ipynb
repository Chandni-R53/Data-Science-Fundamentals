Pyplot functions
- plt.xlabel(), Sets label for x-axis
- plt.ylabel(), Sets label for y-axis
- plt.title(), Sets title of the graph
- plt.grid(), Adds grid lines
- plt.show(), Displays the plot
- plt.legend(), Legend box appears explaining the line(only works if labels are  given in plot)
- plt.xlim(), Sets range for x-axis
- plt.ylim(), Sets range for y-axis
- plt.xticks(), Customize x-axis ticks or rotate labels
- plt.yticks(), Customize y-axis ticks

Line Plot
- plt.plot()

```python
import matplotlib.pyplot as plt
#plt.plot(x,y,color='color name',linestyle='line style',linewidth=value,marker='symbol',label='label name')
x=[5,10,15,20,25,30,35,40]
y=[3,4,7,12,18,9,15,20]
plt.plot(x,y,color="Blue",linewidth=2,linestyle='--',marker='o')
plt.grid(color='orange',linestyle=':',linewidth=1)
plt.show()
```

Scatter Plot
- plt.scatter()
- plt.legend(loc='upper left' or 'lower right')

```python
#plt.scatter(x,y,color='color',marker='symbol',label='label name')
a=[1,2,3,4,5,6,7]
b=[1000,1200,1500,1300,2000,3000,2300]
plt.xlabel("Months")
plt.ylabel("Sales")
plt.title("Monthly Sales Report")
plt.scatter(a,b,color="violet",marker='s',label='Sales Report 2026')
plt.legend(loc='upper left')
plt.show()
```

```python
plt.scatter([1,2,3,4],[56,67,49,63],color='green',marker='o',label='Section A ')
plt.scatter([1,2,3,4],[86,78,94,95],color='orange',marker='o',label='Section B ')
plt.legend()
plt.grid()
plt.show()
```

Bar Plot
- plt.bar()
- plt.barh(), For horizontal bar plot

```python
#plt.bar(x,height,color='color name',width=value,label='label name)
plt.xlim(1,5)
plt.ylim(0,2000)
plt.bar(a,b,color='orange',label='Sales Report 2025')
plt.legend()
plt.xticks([1,2,3,4,5],['M1','M2','M3','M4','M5'])
plt.show()
```

Histogram
- plt.hist()

```python
#plt.hist(data,bins=no.of_bins,color='color name',edgecolor='border color')
scores=[35,39,45,48,57,69,75,75,86,89,94,96,99,92,92,85,78,78,67]
plt.hist(scores,bins=4,color='yellow',edgecolor='blue')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.title('Score Distribution')
plt.show()
```

Pie Chart
- plt.pie()

```python
#plt.pie(values,labels=label_list,colors=color_list,autopct='%1.1f%%')
sub=['Maths','English','Hindi','Science','Sanskrit']
students=[25,18,13,23,16]
plt.title('Subjects Opted by students')
plt.pie(students,labels=sub,autopct='%1.1f%%',colors=['red','yellow','green','blue','purple'])
plt.show()
```
