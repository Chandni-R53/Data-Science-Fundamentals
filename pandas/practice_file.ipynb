## Import Required Libraries

```python
import pandas as pd
import numpy as np
```

## Create a Dataframe

```python

data = {
    "Student_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Aarav", "Diya", "Kabir", "Meera", "Rohan", "Isha"],
    "Department": ["CSE", "ECE", "CSE", "ME", "ECE", "CSE"],
    "Marks_Math": [78, 65, 88, 72, None, 90],
    "Marks_AI": [85, 70, 92, None, 68, 95],
    "Marks_DS": [80, None, 89, 75, 70, 94],
    "Attendance": [92, 85, 96, 88, 80, 98],
    "Exam_Date": ["2024-03-10", "2024-03-10", "2024-03-11", "2024-03-11", "2024-03-12", "2024-03-12"]
}

df = pd.DataFrame(data)
```

## Data Cleaning

```python
df.isna().sum()
```

```python
df.duplicated().sum()
```

```python
df['Exam_Date']=pd.to_datetime(df['Exam_Date'])
```

```python
print(df['Exam_Date'].dtype)
```

```python
df['Marks_Math']=df['Marks_Math'].fillna(df['Marks_Math'].mean())
df['Marks_AI']=df['Marks_AI'].fillna(df['Marks_AI'].mean())
df['Marks_DS']=df['Marks_DS'].fillna(df['Marks_DS'].mean())
```

```python
df.isnull().sum()
```

```python
df.drop(df[df['Attendance']<85].index,inplace=True)
#df=df[df['Attendance']>=85]
```

```python
print(df)
```

## Feature Engineering

```python
df['Total_Marks']=df['Marks_Math']+df['Marks_AI']+df['Marks_DS']
df['Average_Marks']=df['Total_Marks']/3
```

```python
print(df)
```

```python
df['Result']=np.where(df['Average_Marks']>=60,'Pass','Fail')
#np.where(condition, value_if_true, value_if_false)
#df['Result'] = df['Average_Marks'].apply( lambda x: 'Pass' if x >= 60 else 'Fail')
```

```python
print(df['Result'])
```

```python
def grades(x):
    if x>=85:
      return 'A'
    elif x>=70:
      return 'B'
    elif x>=60:
      return 'C'
    else:
      return 'D'

df['Grade']=df['Average_Marks'].apply(grades)
```

```python
df.head()
```

## Advanced Analysis

```python
#department-wise average marks
group=df.groupby('Department')['Average_Marks'].mean()
print(group)
```

```python
#topper of each department
topper=df.loc[df.groupby('Department')['Total_Marks'].idxmax(),['Department','Name']]
print(topper)
```

```python
#attendance-weighted score
df['Weighted_Score']=df['Average_Marks']*(df['Attendance']/100)
print(df['Weighted_Score'])
```

## Reshaping & Filtering

```python
pivot=pd.pivot_table(df,index='Department',columns='Grade',values='Student_ID',aggfunc='count',fill_value=0)
print(pivot)
```

```python
students=df[(df['Grade']=='A')& (df['Attendance']>=90)]['Name']
print(students)
```
