Buitl-in Seaborn Datasets
- tips,iris,flights,penguins etc.

Seaborn Plot Types
- scatterplot(),lineplot(),histplot(),boxplot(),violinplot(),heatmap(),pairplot(),kdeplot(),barplot()

```python
import seaborn as sns
from matplotlib import pyplot as plt
df=sns.load_dataset("flights")
print(df.head(7))
```

Scatter Plot

```python
#sns.scatterplot(x="month",y="passengers",data=df)
sns.scatterplot(x="month",y="passengers",hue="year",size="passengers",sizes=(50,200),data=df)
#hue=for color grouping, size=for marker size
```

Line Plot

```python
sns.lineplot(x="month",y="passengers",data=df)
plt.xlabel("Months")
plt.ylabel("Passengers")
plt.title("Flights Data")
plt.show()
```

```python
#customized line plot
sns.lineplot(x="month",y="passengers",hue="year",marker=True,dashes=False,data=df)
#style=for marker shape, 
#marker=True, it shows markers at each datapoint 
#dashes=False, bydefault dashed line, but now all lines are solid, not dashed.
```

Bar Plot

```python
penguin=sns.load_dataset("penguins")
print(penguin.head())
sns.barplot(x="species",y="body_mass_g",hue="sex", ci=None, palette="pastel",data=penguin)
#palette used for color theme
#ci=none,it removes the confidence interval bars (small vertical line)
plt.xlabel("Species")
plt.ylabel("Body Mass of Species")
plt.title("Bar Plot")
plt.show()
```

Histogram

```python
sns.histplot(x="species",data=penguin,bins=5, kde=True, color="teal")
#bins = number of bars
#kde adds a smooth curve showing the distribution of the data
```

Kernal Density Plot 

```python
sns.kdeplot(x="bill_length_mm",data=penguin,hue="sex", fill=True, alpha=0.6, linewidth=1.5)
#fill used for area under the curve
#alpha controls transparency (opacity) of the color
```

Box Plot

```python
sns.boxplot(x="species",y="body_mass_g",data=penguin,hue="sex", palette="Set2", linewidth=1.5, fliersize=8)
#fliersize changes the size of outlier dots
```

Violin Plot

```python
sns.violinplot(x="species",y="body_mass_g",data=penguin)
```

Pair Plot

```python
sns.pairplot(data=penguin)
sns.pairplot(data=penguin,hue="species", diag_kind="kde", palette="husl")
```

Heatmap

```python
tips_dataset=sns.load_dataset("tips")
corr=tips_dataset.corr(numeric_only=True)
sns.heatmap(corr)
```
